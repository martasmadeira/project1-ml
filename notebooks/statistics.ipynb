```python
import numpy as np
import pandas as pd
from scipy.io import loadmat

import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.feature_extraction.text import CountVectorizer
```

```python
df = pd.read_csv("C:\\Users\\carlo\\OneDrive - Danmarks Tekniske Universitet\\1st Year 1st Semester\\02452 Machine Learning\\project1-ml\\ObesityDataSet_raw_and_data_sinthetic.csv")
 
df.head()
print('shape:',df.shape)
print(df.info())
print('columns:',df.columns)
df.head()
```

```python
df.hist(figsize=(18, 14), bins=20, edgecolor='black')
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5)) 

df["Gender"].value_counts().plot(kind="bar", edgecolor="black",ax=axes[0],  title="Gender Distribution")

order = pd.CategoricalDtype(categories=["no", "Sometimes", "Frequently", "Always"], ordered=True)
df["CAEC"] = df["CAEC"].astype(order)
df["CAEC"].value_counts().sort_index().plot(kind="bar", edgecolor="black",ax=axes[1], title="Food between meals (CAEC)")

df["NObeyesdad"].value_counts().plot(kind="barh", edgecolor="black",ax=axes[2], title="Obesity Levels")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

num_cols = ["Age", "Height", "Weight", "NCP", "CH2O", "FAF", "TUE"]

plt.figure(figsize=(14, 6))
df[num_cols].boxplot()
plt.title("Boxplots of Continuous Features")
plt.xticks(rotation=45)
plt.show()

# ok lol 
```

```python
#correlation matriz 
plt.figure(figsize=(10, 8))
correlation_matrix = df.corr(numeric_only=True)
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", square=True)
plt.title("Correlation Matrix")
```

```python
num_cols = ["Age", "Height", "Weight", "NCP", "CH2O", "FAF", "TUE"]

# pairplot com todas as numéricas, colorindo pelas classes de obesidade
sns.pairplot(df[num_cols + ["NObeyesdad"]], hue="NObeyesdad", diag_kind="hist", markers="o")

plt.suptitle("Pairplot of Numerical Features in the Obesity Dataset", y=1.02)
plt.show()
```
